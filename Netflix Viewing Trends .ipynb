## 1.0.6 📊 Analyzing and Visualizing data


## 🎬 1. Distribution of Content Type (Movies vs TV Shows)

This plot helps us understand the proportion of movies and TV shows available on Netflix.

```python
sns.countplot(x='type', data=df)
plt.title('Distribution of Content Type on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## 🌍 2. Top 10 Countries Producing Netflix Content

We analyze the countries contributing the most content to Netflix's library.

```python
top_countries = df['country'].value_counts().head(10)
print("Top 10 countries producing Netflix content:\n")
print(top_countries)
```

```python
top_countries.plot.barh()
plt.title("Top 10 countries producing Netflix content")
plt.show()
```

## 📅 3. Content Added by Year

This plot shows how Netflix’s content library has grown over time, based on the year titles were added.

```python

df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(12, 6))
df['year_added'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Titles Added to Netflix by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

## 🎭 4. Top 10 Most Common Genres

Netflix content is categorized into multiple genres. This chart shows which genres are most frequently represented.

```python
genres = df['listed_in'].dropna().str.split(', ').explode()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

## 🕐 5. Movie Duration Distribution

We examine how long Netflix movies typically are, based on the duration in minutes.

```python
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration'] = movies_df['duration'].str.extract(r'(\d+)').astype(float)

plt.figure(figsize=(10, 5))
sns.histplot(movies_df['duration'], bins=30, kde=True)
plt.title('Distribution of Movie Durations (in minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

## 🥧 5. Distribution of Content Ratings on Netflix

Ratings help viewers understand the age-appropriateness of shows and movies.
Let's visualize the distribution of the most common ratings to see what kind of content dominates Netflix.

```python
# Get the top 6 most common ratings
rating_counts = df['rating'].value_counts().head(6)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', colors=sns.color_palette("Set3"))
plt.title('Top 6 Ratings Distribution on Netflix')
plt.show()
```

## Multiple Feature Plots

This visualization shows how Netflix’s top 3 genres are distributed across ratings and countries.
It helps identify which genres dominate specific content ratings and popular regions.

```python
# Get top 3 genres
top_genres = df['listed_in'].str.split(', ').explode().value_counts().head(3).index

# Select features to compare
features = ['rating', 'country']

plt.figure(figsize=(12, 8))

for i, feature in enumerate(features):
    plt.subplot(1, len(features), i+1)
    top_data = df[df['listed_in'].str.contains('|'.join(top_genres), na=False)]
    sns.countplot(data=top_data, x=feature, order=top_data[feature].value_counts().index[:6], hue=top_data['listed_in'].str.split(', ').str[0])
    plt.title(f'{feature} Distribution by Genre')
    plt.xticks(rotation=45)
    plt.xlabel(feature)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

## Conclusion

The analysis of the Netflix dataset reveals several key insights. 
Netflix primarily offers Movies over TV Shows, with a significant rise in content additions leading up to 2020,
followed by a slight decline. The United States is the top contributor to the platform’s content, followed by India, the United Kingdom, and Canada.
July and December are the most common months for content releases. In terms of ratings, TV-MA, TV-14, and PG-13 dominate, especially in Movies,
while kid-friendly ratings like TV-Y and TV-G are more common in TV Shows. Most movies have a typical duration between 80 and 100 minutes
Dramas, Comedies, and Documentaries are the most prevalent genres across the platform. Visual analysis also shows that Dramas and Documentaries
dominate Movies, while genres like Kids’ Shows are more frequent in TV Shows. Country-wise genre distribution varies, reflecting local audience
preferences. Lastly, multi-feature comparisons—such as genre vs. type and genre vs. rating—revealed interesting patterns in content distribution,
highlighting how certain genres are more associated with specific content types and audience ratings.
